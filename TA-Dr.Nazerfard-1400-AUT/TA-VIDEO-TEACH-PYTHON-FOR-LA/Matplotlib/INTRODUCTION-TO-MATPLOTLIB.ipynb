## plot

```python
plt.plot([1, 2, 3])
```

```python
import numpy as np
x = np.arange(1, 20)
y = np.sin(x)
plt.plot(x, y)
```

```python
plt.plot(x, y, linestyle="solid")
```

```python
plt.plot(x, y, linestyle="dashed")
```

```python
plt.plot(x, y, linestyle="solid", marker="*")
```

```python
plt.plot(x, y, linestyle="solid", marker=".")
```

```python
plt.plot(x, y, color="red")
```

```python
plt.plot(x, y, "go", markersize=5)
```

```python
plt.plot(x, y)
plt.xlim(0, 100)
plt.ylim(1, 50)
plt.title("y = sin(x)")
```

```python
plt.plot(x, y)
plt.xlabel("X")
plt.ylabel("sin X")
```

```python
plt.plot(x, y)
plt.plot(x, y* 2)
plt.legend(["y", "2y"])
```

## scatter

```python
plt.scatter(x, 2 * y)
plt.plot(x, y)
```

## subplot

```python

plt.figure(figsize=(6, 3))
plt.subplot(2, 2, 1)
plt.plot(x, y)
plt.subplot(2, 2, 2)
plt.scatter(x, y)

plt.subplot(2, 2, 3)
plt.plot(x, 2*y, color = "red")
plt.subplot(2, 2, 4)
plt.scatter(x, 2*y, color = "red")
```
